```python
import pandas as pd

from sentence_transformers import SentenceTransformer

from sklearn.cluster import KMeans
```

```python
model = SentenceTransformer('jhgan/ko-sroberta-multitask')
```

```python
df = pd.read_csv('../data/eldenring_stream_reviews_ko.csv')

def preprocessing(df):
    new_df = df.copy()
    new_df['review'] = new_df['review'] \
      .replace(r'[^가-힣 ]', ' ', regex=True) \
      .replace("'", '') \
      .replace(r'\s+', ' ', regex=True) \
      .str.strip() \
      .str[:255]
    new_df = new_df[new_df['review'].str.strip().astype(bool)]
    new_df['review'].values.tolist()[:5]
    return new_df

df = preprocessing(df)

corpus = df['review'].values.tolist()

embeddings = model.encode(corpus)
```

```python
num_clusters = 3
clustering_model = KMeans(n_clusters=num_clusters)
clustering_model.fit(embeddings)
cluster_assignment = clustering_model.labels_

clustered_sentences = {i: [] for i in range(num_clusters)}
for sentence_id, cluster_id in enumerate(cluster_assignment):
    clustered_sentences[cluster_id].append(corpus[sentence_id])
```
